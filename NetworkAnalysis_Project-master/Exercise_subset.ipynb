```python
import networkx as nx
import matplotlib.pyplot as plt

def generator(nodes, exponent):
    while True:  
        s=[]
        while len(s)<nodes:
            nextval = int(nx.utils.powerlaw_sequence(1, exponent)[0])
            if nextval!=0:
                s.append(nextval)
        if sum(s)%2 == 0:
            break
    G = nx.configuration_model(s)
    G = nx.Graph(G) # remove parallel edges
    G.remove_edges_from(nx.selfloop_edges(G))
    return G

exponent_2_2 = generator(10000,2.2)
exponent_2_5 = generator(10000,2.5)
exponent_3_0 = generator(10000,3)
```

```python
pos = nx.spring_layout(exponent_2_2)
nx.draw_networkx(exponent_2_2, pos)
plt.show()
```
